### <b>Question No. 1</b>

Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in Python. It provides a wide variety of plots and charts for visualizing data in a clear and concise manner.<br>
The five plots that can be plotted using the Pyplot module of Matplotlib are:

1. **Line Plot**: Used to visualize data points as a series of connected markers, typically used to show trends over time.
2. **Bar Chart**: Used to compare categories of data using rectangular bars, where the length of each bar represents the value of the data.
3. **Histogram**: Used to visualize the distribution of a dataset, showing the frequency of values within specified bins.
4. **Scatter Plot**: Used to visualize the relationship between two variables by plotting points on a two-dimensional plane.
5. **Pie Chart**: Used to represent data in a circular graph, where each slice represents a proportion of the whole.

### <b>Question No. 2

A scatter plot is a type of plot that displays values for two variables as points on a two-dimensional plane, where each point represents an observation in the dataset. Scatter plots are useful for visualizing the relationship between two continuous variables.

Here's how you can generate the data for `x` and `y` using the provided code and plot a scatter plot:

In this code, `plt.scatter` is used to create a scatter plot of `x` and `y` values. The `title`, `xlabel`, and `ylabel` functions are used to add a title and labels to the plot. The `grid` function is used to add a grid to the plot for better readability.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for x and y
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plot a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', alpha=0.5)
plt.title('Scatter Plot of x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

### <b>Question No. 3

The `subplot()` function in Matplotlib is used to create multiple plots (subplots) within the same figure. It allows you to arrange plots in a grid-like layout, specifying the number of rows and columns of subplots.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot each line plot on a separate subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

### <b>Question No. 4

A bar plot is a type of plot that represents categorical data with rectangular bars, where the length or height of each bar corresponds to the value of the data it represents. Bar plots are commonly used to compare the values of different categories or to show the distribution of a single categorical variable.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Bar Plot of Profit by Company')
plt.xticks(rotation=45)
plt.show()

# Create a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='lightcoral')
plt.ylabel('Company')
plt.xlabel('Profit')
plt.title('Horizontal Bar Plot of Profit by Company')
plt.show()
```

### <b>Question No. 5

A box plot, also known as a box-and-whisker plot, is a type of plot that displays the distribution of a dataset along a single axis, showing the median, quartiles, and potential outliers. It is used to summarize the distribution of a dataset and to compare the distribution of different datasets.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Value')
plt.title('Box Plot of Box 1 and Box 2')
plt.grid(True)
plt.show()
```
